- 정량적 지표로는 Rouge1 및 Rouge2를 사용하였습니다. 
- 정성적 지표로는 사람이 작성한 예측문을 활용하여 모델의 예측 결과를 평가합니다.

```python
score_lst = ['precision_1', 'recall_1', 'fmeasure_1', 'precision_2', 'recall_2', 'fmeasure_2']

def split_scores(x):
    lst = x.split(',')
    
    #- rouge1
    precision_1 = lst[0].split('=')[1]
    recall_1 = lst[1].split('=')[1]
    fmeasure_1 = lst[2].split('=')[1][:-1]
    
    #- rouge2
    precision_2 = lst[3].split('=')[1]
    recall_2 = lst[4].split('=')[1]
    fmeasure_2 = lst[5].split('=')[1][:-2]
    
    return precision_1, recall_1, fmeasure_1, precision_2, recall_2, fmeasure_2

def exe_split_scores(df):
    
    for i,j in enumerate(score_lst):
        df[j] = df['scores'].apply(lambda x: round(float(split_scores(x)[i]),2))
        
    return df

def show_data(df, ext=True, step=10 , threshold=10 ):
    cnt = 0
    for i in range(1, df.shape[0], step):
        if ext:
            sum_en = 'ext'; sum_ko = '추출요약'
        else: 
            sum_en = 'abs'; sum_ko = '생성요약'
        
        print("{}번째 원문\n{}\n\n{}번째 {}\n{}\n\n{}번째 예측\n{}\n"\
              .format(i, df['original'][i], i, sum_ko, df[sum_en][i], i, df['preds'][i]))
        
        if cnt > threshold: break 
        cnt += 1
```

```python
show_data(pred_ext, ext=True, step=10 , threshold=10 )
```

```python
show_data(pred_abs, ext=False, step=10 , threshold=10 )
```

```python
pred_ext = exe_split_scores(pred_ext)
pred_abs = exe_split_scores(pred_abs)
```

```python
pred_abs.rename(columns={'preds':'pred_abs'}, inplace=True)
pred_ext.rename(columns={'preds':'pred_ext'}, inplace=True)

inter = ['original','abs','ext']
abs_ = ['pred_abs'] + inter + score_lst
ext = ['pred_ext'] + inter + score_lst
```
